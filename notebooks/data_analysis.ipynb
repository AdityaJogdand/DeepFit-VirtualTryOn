```python
import os
import shutil

IMAGE_DIR = "../images"
SEGM_DIR = "../segm"

NEW_IMAGE_DIR = "../dataset_clean/images"
NEW_SEGM_DIR = "../dataset_clean/segm"

os.makedirs(NEW_IMAGE_DIR, exist_ok=True)
os.makedirs(NEW_SEGM_DIR, exist_ok=True)

images = os.listdir(IMAGE_DIR)
segm = set(os.listdir(SEGM_DIR))

count = 0

for img in images:

    segm_name = img.replace(".jpg", "_segm.png")

    if segm_name in segm:

        shutil.copy(
            os.path.join(IMAGE_DIR, img),
            os.path.join(NEW_IMAGE_DIR, img)
        )

        shutil.copy(
            os.path.join(SEGM_DIR, segm_name),
            os.path.join(NEW_SEGM_DIR, segm_name)
        )

        count += 1

print("Total labeled samples copied:", count)
```

```python
import os
import json
import shutil

# original dataset
IMAGE_DIR = "../images"
SEGM_DIR = "../segm"

# clean dataset folders
CLEAN_IMAGE_DIR = "../dataset_clean/images"
CLEAN_SEGM_DIR = "../dataset_clean/segm"

os.makedirs(CLEAN_IMAGE_DIR, exist_ok=True)
os.makedirs(CLEAN_SEGM_DIR, exist_ok=True)

images = [f for f in os.listdir(IMAGE_DIR) if f.endswith(".jpg")]
segm = set(os.listdir(SEGM_DIR))

dataset = []
copied = 0

for img in images:

    segm_name = img.replace(".jpg", "_segm.png")

    if segm_name in segm:

        image_src = os.path.join(IMAGE_DIR, img)
        segm_src = os.path.join(SEGM_DIR, segm_name)

        image_dst = os.path.join(CLEAN_IMAGE_DIR, img)
        segm_dst = os.path.join(CLEAN_SEGM_DIR, segm_name)

        # copy files
        shutil.copy(image_src, image_dst)
        shutil.copy(segm_src, segm_dst)

        sample = {
            "image": image_dst,
            "segm": segm_dst
        }

        dataset.append(sample)
        copied += 1

print("Total labeled samples:", copied)

# save dataset index
with open("../dataset_clean/dataset_index.json", "w") as f:
    json.dump(dataset, f, indent=4)

print("Dataset index saved to dataset_clean/dataset_index.json")
```
